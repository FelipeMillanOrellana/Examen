```python
import numpy as np

entradas_platinum = np.empty(20, object)
entradas_gold = np.empty(30, object)
entradas_silver = np.empty(50, object)
asistentes = []

def menu():
    while True:
        print("Bienvenidos a la pagina ofcila de Creativos")
        print("-"*45)
        print("1. Comprar entradas")
        print("2. Mostrar ubicaciones disponibles")
        print("3. Ver listado de asistentes")
        print("4. Mostrar ganancias totales")
        print("5. Salir")
        opcion = input("Ingrese una opción: ")

        if opcion == "1":
            comprar_entradas()
        elif opcion == "2":
            mostrar_ubicaciones_disponibles()
        elif opcion == "3":
            ver_listado_asistentes()
        elif opcion == "4":
            mostrar_ganancias_totales()
        elif opcion == "5":
            nombre = input(str("Ingrese su nombre: "))
            apellido = input(str("Ingrese su apellido: "))
            fecha_actual = obtener_fecha_actual()
            print(f"Gracias por utilizar el sistema, {nombre} {apellido}. ¡Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, ingrese una opción válida.")

def comprar_entradas():
    print("Comprar entradas")
    cantidad = input("Ingrese la cantidad de entradas a comprar : ")

    if cantidad.isdigit() and 1 <= int(cantidad) <= 3:
        cantidad = int(cantidad)
        mostrar_ubicaciones()
        seleccionar_ubicaciones(cantidad)
    else:
        print("Cantidad inválida. Debe ingresar un número entre 1 y 3.")

def mostrar_ubicaciones():
    print("Ubicaciones disponibles:")
    print("Platinum:", np.count_nonzero(entradas_platinum == None))
    print("Gold:", np.count_nonzero(entradas_gold == None))
    print("Silver:", np.count_nonzero(entradas_silver == None))

def seleccionar_ubicaciones(cantida):
    for i in range(cantida):
        ubicacion = input(f"Ingrese la ubicación {i + 1}: ")

        if ubicacion.isdigit():
            ubicacion = int(ubicacion)

            if ubicacion >=1 and ubicacion <= 20:
                if entradas_platinum[ubicacion - 1] is None:
                    run = input("Ingrese el RUN (sin puntos ni guiones): ")

                    if run.isdigit():
                        asistentes.append({"run": run, "ubicacion": f"Platinum-{ubicacion}"})
                        entradas_platinum[ubicacion - 1] = run
                        print("Operación realizada correctamente.")
                    else:
                        print("RUN inválido. Debe ingresar un número.")
                else:
                    print("Ubicación no disponible.")
            else:
                print("Ubicación inválida. Debe ingresar un número entre 1 y 20.")
        elif ubicacion.isdigit():
            ubicacion = int(ubicacion)

            if ubicacion>=21 and ubicación <= 50:
                if entradas_gold[ubicacion - 21] is None:
                    run = input("Ingrese el RUN (sin puntos ni guiones): ")

                    if run.isdigit():
                        asistentes.append({"run": run, "ubicacion": f"Gold-{ubicacion}"})
                        entradas_gold[ubicacion - 21] = run
                        print("Operación realizada correctamente.")
                    else:
                        print("RUN inválido. Debe ingresar un número.")
                else:
                    print("Ubicación no disponible.")
            else:
                print("Ubicación inválida. Debe ingresar un número entre 21 y 50.")
        elif ubicacion.isdigit():
            ubicacion = int(ubicacion)

            if ubicacion>= 51 and ubicacion <= 100:
                if entradas_silver[ubicacion - 51] is None:
                    run = input("Ingrese el RUN (sin puntos ni guiones): ")

                    if run.isdigit():
                        asistentes.append({"run": run, "ubicacion": f"Silver-{ubicacion}"})
                        entradas_silver[ubicacion - 51] = run
                        print("Operación realizada correctamente.")
                    else:
                        print("RUN inválido. Debe ingresar un número.")
                else:
                    print("Ubicación no disponible.")
            else:
                print("Ubicación inválida, debe ingresar un número entre 51 y 100.")
        else:
            print("Ubicación inválida, debe ingresar un número.")

def mostrar_ubicaciones_disponibles():
    print("Estado actual de la venta de entradas:")
    print("Platinum:", entradas_platinum)
    print("Gold:", entradas_gold)
    print("Silver:", entradas_silver)

def ver_listado_asistentes():
    print("Listado de asistentes:")
    asistentes_ordenados = sorted(asistentes, key=lambda x: int(x["run"]))
    for asistente in asistentes_ordenados:
        print(f"RUN: {asistente['run']}, Ubicación: {asistente['ubicacion']}")

def mostrar_ganancias_totales():
    print("Total ganado por concepto de entradas:")
    total_platinum = 120000 * np.count_nonzero(entradas_platinum == None)
    total_gold = 80000 * np.count_nonzero(entradas_gold == None)
    total_silver = 50000 * np.count_nonzero(entradas_silver == None)
    total = total_platinum + total_gold + total_silver

    print("Tipo Entrada\tCantidad\tTotal")
    print(f"Platinum\t{np.count_nonzero(entradas_platinum == None)}\t\t${total_platinum}")
    print(f"Gold\t\t{np.count_nonzero(entradas_gold == None)}\t\t${total_gold}")
    print(f"Silver\t\t{np.count_nonzero(entradas_silver == None)}\t\t${total_silver}")
    print(f"TOTAL\t\t{np.count_nonzero(entradas_platinum == None) + np.count_nonzero(entradas_gold == None) + np.count_nonzero(entradas_silver == None)}\t\t${total}")
menu()
```
